# The time_variation routine in striptease

This procedure analyzes the scientific time string, which looks for the jumps and falls in time vector in order to classify them into three type: forwards, backwards and hats. 
The main result is a dictionary with information (index and time) for each polarimeter.


## Importing Module and data:

For illustration, we analyze a test focus on effects of parameter settings on time vectors: phase_src and clk_ref. These test are available in /mnt/storage/data/HDF5/2020/04/ and we use the 2020_04_01_14-31-52.h5, please, use the correct *path* for you computer.

```python
from striptease import time_variation as tvar 
path ='/Users/clopez/work_strip/dataStrip/labTest/' # my path
file =path+'2020_04_01_14-31-52.h5'
```

We use the timevariation class to run the analysis:
